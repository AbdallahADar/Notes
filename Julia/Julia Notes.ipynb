### Arrays:

```julia
#Initializing a simple 1D array

A = [1,2,3,4,5,6] #1D list
println(A)
println(A[1]," First Element") #Indexing starts from 1
println(A[end]," Last Element") #We can use end to get last element of an array
println(A[end-1]," Second Last Element") #We can use end as reference  to get second last or thrid last element or etc of an array
println(length(A), " Length of Array") #To get length
println(A[length(A)]," Using length to get last element") #To get last element
println(A[:], " Get all elements") #Get all elements
println(A[1:2], " Slicing array") #Slicing: Includes starting and ending index so [1:2] means item at index 1 and index 2
println(A[1:2:length(A)]," Slicing with step size") #Slicing: Start,step,end
println(A[:2:5]," ", A[2:5]," If we have no number before the first colon, it is disregarded. A[:x:y] is same as A[x:y]") #First colon without a preceding number is ignored
println(A[length(A):-1:1]," Reversing array") #Slicing: Reversing the list
println(A[[3,5]]," Indexing multiple elements") #To get multiple elements at position we use double square brackets. Essentially a vector as the indexing input. A[[x,y]] is A[l] where l=[x,y]
println(A[[true,false,true,true,false,true]], " We can also use boolean indexing") #To get boolean indexing we use double square brackets with array being a boolean array
```

### Tuples:

```julia
#Tuples are similar to arrays but are declared using parenthesis.
#Tuples are immutable. That is their values cannot be changed after initialization. This is a key difference compared to arrays.
```

```julia
#The declaration is difference but indexing rules are same as an array

T1 = (1,2,3,4,5,6)
println(T1," Tuple")
println(T1[1]," First Element") #Indexing starts from 1
println(T1[end]," Last Element") #We can use end to get last element of an array
println(T1[end-1]," Second Last Element") #We can use end as reference  to get second last or thrid last element or etc of an array
println(length(T1), " Length of Array") #To get length
println(T1[length(T1)]," Using length to get last element") #To get last element
println(T1[:], " Get all elements") #Get all elements
println(T1[1:2], " Slicing array") #Slicing: Includes starting and ending index so [1:2] means item at index 1 and index 2
println(T1[1:2:length(T1)]," Slicing with step size") #Slicing: Start,step,end
println(T1[:2:5]," ", T1[2:5]," If we have no number before the first colon, it is disregarded. A[:x:y] is same as A[x:y]") #First colon without a preceding number is ignored
println(T1[length(T1):-1:1]," Reversing array") #Slicing: Reversing the list
println(T1[[3,5]]," Indexing multiple elements") #To get multiple elements at position we use double square brackets. Essentially a vector as the indexing input. A[[x,y]] is A[l] where l=[x,y]
println(T1[[true,false,true,true,false,true]], " We can also use boolean indexing") #To get boolean indexing we use double square brackets with array being a boolean array
```

## Loops:

```julia
#For loops on array. Same for tuple. Same for set
for i in ["A","B","C"]
    println(i*":")
end
```

```julia
#Note: Remember that a string is also a container that we can iterate over using a for loop
```

```julia
#For loops on dict
for i in Dict(:A=>20,:B=>40)
    println(i) # Pair of key and value. A pair of symbol and int in this case
    println(i[1]," ",i[2]) #Key is at first index. Value is at second index
end
```

```julia
#For loop on range
#Recall when talking about range we used collect. Collect converted range to 1D array. When iterating, we do not need to convert it. We can use it directly
for i in 0:2:10 #Start:Step:End
    println(i)
end
```

```julia
#For loop on range collected into 1d array. Gives same output
for i in collect(0:2:10) #Start:Step:End
    println(i)
end
```

```julia
#Using print trick for easy printing
for i in 1:10
    j = i*10
    println("$(j) is multiplication between $(i) and 10") #A very easy print trick. Use $(var_name) to display variables that are already defined.
end
```

```julia
#Using continue
for i in 1:10
    if i%2 == 0
        continue #Continue keyword states that we should stop here and move to next iteration
    end
    println(i)
end
```

```julia
#Pairs in for loops is just nested loops
for x in 1:2, y in 3:4
    println(x," ",y)
end
```

```julia
#Triplets in for loops is triple nested loops
for x in 1:2, y in 3:4, z in 5:6
    println(x," ",y," ",z)
end
```

```julia
#While loops
a = 1
while a < 5
    println(a)
    a += 1
end
```

```julia
#Creating simple scatter plot
scatter(x,y,title="TITLE", xlabel="X Added",ylabel = "Y Added")
#the label,title axis labels discussed above hold here as well
```

```julia
#Multiple scatter plots
scatter(x,y,label="A")
scatter!(x,z,label="B")
```

```julia
#Simple bar plot
#Note this bar plot has y in the y axis and x in x axis. Its not a frequency or probability plot. Dont confuse with histogram
#A way to think about this is that this would should a frequency or probability plot if y was the frequency/probability
bar(x,y,title="TITLE", xlabel="X Added",ylabel = "Y Added")
#the label,title axis labels discussed above hold here as well
```

```julia
#Multiple scatter plots
#We introduce alpha. alpha is a measure of opaqueness (translusency) and is useful to change when graphs are overlapping each other
bar(x,y,label="A",alpha=0.75)
bar!(x,z,label="B",alpha=0.75)
```

```julia
#Simple histogram
#Shows count on y axis
#bins parameter is to choose number of bins
histogram(rand(15),title="TITLE",bins=10)
#the label,title axis labels discussed above hold here as well
```

```julia
#Multiple Histograms
#We can have different sized bins
histogram(rand(15),alpha = 0.75,bins=10)
histogram!(rand(15),alpha = 0.75,bins=5)
```

```julia
#Multiple Plots / Grid of plots

#Simple line plot
p1 = plot(x,y,title="LINE") 

#Multiple scatter plot
p2 = scatter(x,y)
p2 = scatter!(p2,x,z,xlabel = "X-axis",ylabel = "Y_axis")

#Bar Plot
p3 = bar(y,bins=10)

#Multiple histogram plot
p4 = histogram(y,alpha = 0.75,bins=10)
p4 = histogram!(p4, z ,alpha = 0.75,bins=15)

plot(p1,p2,p3,p4, layout = (2,2))
```

```julia
#We can remove legend for all by simply choosing legend = false in plot
plot(p1,p2,p3,p4, layout = (2,2), legend = false)
```

```julia
#Setting title will set the title to each graph in layout, same goes for x and ylabel. In short anything in the plot argument will be for all graphs in the grid
plot(p1,p2,p3,p4, layout = (2,2), legend = false,title="Overwrite",xlabel="X-axis",ylabel="Y-axis")
```

```julia
#Lets combine scatter and line plot and add it to the grid
p5 = plot(x,y)
p5 = scatter!(p5,x,y)

#Add to grid
#We are allowed to define a layout that may have more grid cells than plots we provide
plot(p1,p2,p3,p4,p5, layout = (2,3), legend = false)
```

```julia
#We are not allowed to define a layout that may have less grid cells than plots we provide
plot(p1,p2,p3,p4,p5, layout = (2,2), legend = false)
```
