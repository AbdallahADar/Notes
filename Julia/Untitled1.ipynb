```python
import numpy as np
import pandas as pd
```

```python
frequents = pd.DataFrame()
frequents["drinker"] = ["A","B","C","A","D","B"]
frequents["bar"] = [1,1,2,2,3,3]

sells = pd.DataFrame()
sells["beer"] = ["V","X","Y","Z","V","X","Z"]
sells["bar"] = [1,1,1,1,2,2,3]

likes = pd.DataFrame()
likes["drinker"] = ["A","B","C","D","E","A","E","D","B","C"]
likes["beer"] = ["V","V","V","V","V","Z","Z","X","Y","Y"]
```

```python
frequents
```

```python
sells
```

```python
likes
```

```python
'''
select * from sells a
LEFT JOIN frequents b
on a.bar=b.bar
'''
combo1 = sells.merge(frequents,how="left",on="bar")
combo1
```

```python
'''
select distinct(drinker) from
(
select * from likes l
left join 
(
select * from sells a
LEFT JOIN frequents b
on a.bar=b.bar
) t
on l.drinker=t.drinker and l.beer = t.beer
)
where bar is not null
'''
likes.merge(combo1,how="left",on=("drinker","beer"))
```

```python
import numpy as np
import pandas as pd
from datetime import datetime
from sklearn.decomposition import PCA
```

```python
fullYCweb = pd.read_html("https://goo.gl/j97141")
```

```python
YCdata = fullYCweb[0]
```

```python
YCdata['Date'] =YCdata['Date'].astype('datetime64')
YCdata
```

```python
df = pd.concat([YCdata["Date"],YCdata.iloc[:,14:21],YCdata["30 Yr"]],axis=1)
```

```python
df.rename(columns={'1 mo': 1/12,
'3 mo': 1/4, '6 mo': 1/2, '1 yr': 1,
'2 yr': 2, '3 yr': 3, '5 yr': 5, '7 yr': 7, '10 yr': 10, '20 yr': 20, '30 yr': 30}, inplace=True)
```

```python
pcaout = PCA(n_components=8)
```

```python
pcaout.fit(df.drop('Date', axis=1))
```

```python
pcaout.explained_variance_ratio_
```

```python
pcaout.components_[0]
```

```python
pcaout.explained_variance_ #Eigenvalues
```

```python
newshift = np.array([0.01,
             0.01, 0.07, 0.05, 0.05, 0.03, 0.03, 0.04])
pcaout.transform(newshift.reshape(1, -1))
```

```python
newshift.reshape(1, -1)@pcaout.components_
```

```python
pcaout.components_@newshift.reshape(-1, 1)
```

```python
newshift.reshape(1, -1) - pcaout.mean_ @ pcaout.components_.T
```
